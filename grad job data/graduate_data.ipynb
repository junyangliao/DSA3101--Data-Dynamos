```python
def left_join_ksa(df1, ksa):
    merged = pd.merge(df1, ksa, on='O*NET-SOC Code', how = 'left').drop(columns = ["Title_y", "Element ID"]).rename(columns={"Element Name":"Skills", "Title_x":"Title"})
    return merged
```
